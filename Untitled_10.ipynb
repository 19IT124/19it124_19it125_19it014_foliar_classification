```
import numpy as np
import matplotlib.pyplot as plt
import numpy as np
import os
import cv2
from numpy import asarray
```

```
DataDir = "Train/"
Class_Name = ["scab","rush","multi-diseases","healthy"]
print(len(Class_Name))

for clas in Class_Name:
    path = os.path.join(DataDir,clas)
    for img in os.listdir(path):
        img_array = np.array(cv2.imread(os.path.join(path,img))) #, cv2.IMREAD_GRAYSCALE
        plt.imshow(img_array, cmap="gray")
        plt.show()
        print(img_array.shape)
        print(type(img_array))
        break
    break
```

```
training_data = []
Traning_Images = []
Traning_Labels = []
img_size = 120

def cerate_training_data():
    count = 0
    for clas in Class_Name:
        path = os.path.join(DataDir,clas)
        class_num = Class_Name.index(clas)
        for img in os.listdir(path):
            #try:
            img_array = cv2.imread(os.path.join(path,img))#, cv2.IMREAD_GRAYSCALE
            img_array = cv2.cvtColor(img_array, cv2.COLOR_BGR2RGB)
            data = asarray(img_array)
            #plt.imshow(new_array, cmap="gray")
            #plt.show()
            print(img)
            print("Train image: ",count)
            training_data.append([data, class_num])
            Traning_Images.append(data)
            Traning_Labels.append(asarray(class_num))
            #img_array2 = cv2.imread(os.path.join(path,img))#, cv2.IMREAD_GRAYSCALE
            #new_array2 = cv2.resize(img_array, (img_size,img_size))
            #Traning_ImageC.append(np.array([img_array2]))
            #except Exception as e:
            #    print (e)
            count = count + 1
        
cerate_training_data()
```

```
np.save("Traning_ImagesLD",np.array(Traning_Images))
np.save("Traning_LabelsLD",np.array(Traning_Labels))
```
